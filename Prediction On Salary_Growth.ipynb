Q9. description of minimum, mean and maximum Salary on the basis of position?

```python
Employee_salary_by_position = pd.DataFrame(df1.groupby('Position') 
                            .agg(min_sal = ('Salary', 'min'), 
                                 mean_sal = ('Salary', 'mean'),
                                 max_sal = ('Salary', 'max')).round(2))
Employee_salary_by_position
```
